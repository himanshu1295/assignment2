Parse file using json,csv python library and write output to csv file.

```python
# importing json and csv library

import json
import csv

# opening the JSON file
with open('sample_output.json') as f:
    data = json.load(f)

# creating a CSV writer object
csv_writer = csv.writer(open('assignment_2.csv', 'w'))

# writing the header row
csv_writer.writerow(data[0].keys())

# writing the data rows
for row in data:
    csv_writer.writerow(row.values())
```
